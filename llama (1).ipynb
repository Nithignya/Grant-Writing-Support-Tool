```python
from langchain.llms import HuggingFacePipeline
```

```python
import transformers
```

```python
import torch
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# model="meta-llama/Llama-2-7b-chat-hf"
model="daryl149/llama-2-7b-chat-hf"
```

```python
from transformers import AutoTokenizer
```

```python
tokenizer=AutoTokenizer.from_pretrained(model)
```

```python
pipeline=transformers.pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map="auto",
    max_length=1000,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id
)
```

```python
llm=HuggingFacePipeline(pipeline=pipeline, model_kwargs={'temperature':0})
```

```python
prompt="Write a grant proposal for a funding of Rupees 50,000 for a project to take care of Street animals."
```

```python
print(llm(prompt))
```

```python
from huggingface_hub import hf_hub_download
from llama_cpp import Llama
```

```python
model_path = hf_hub_download(repo_id=model_name_or_path, filename=model_basename)
```

```python
print(model_path)
```

```python
lcpp_llm = None
lcpp_llm = Llama(
    model_path = model_path,
    n_threads=2,
    n_batch=512,
)
```

```python
prompt= "Write a linear regression code"
prompt_template=f'''SYSTEM: You are a helpful, respectful and honest assistant. You always answer helpfully.

user: {prompt}
ASSISTANT:
'''
```

```python
respose=lcpp_llm(prompt=prompt_template, max_tokens=256, temperature=0.5,top_p=0.95, repeat_penalty=1.2, top_k=150, echo=True)
```

```python
print(respose)
```

```python
print(respose["choices"][0]["text"])
```
